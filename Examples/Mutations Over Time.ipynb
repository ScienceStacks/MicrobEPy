## Bar Plots of Raw Data

```python
mutation_plot = MutationLinePlot(mutation_column=cn.GGENE_ID)
parms = PlotParms()
parms[cn.PLT_FIGSIZE] = [20, 16]
mutation_plot.plotTransfers(parms=parms)
```

## Correlations of Mutations

```python
parms = PlotParms()
parms[cn.PLT_TITLE] = str(cn.TRANSFER_DEFAULT)
parms[cn.PLT_FIGSIZE] = (16, 12)
mutation_plot = MutationLinePlot(mutation_column=cn.GGENE_ID)
mutation_plot.plotSiglvl(parms=parms, max_siglvl=0.05)
```

```python
parms = PlotParms()
parms[cn.PLT_FIGSIZE] = (20, 16)
mutation_plot.plotSiglvls(max_siglvl=0.05, parms=parms)
```

```python
parms = PlotParms()
parms[cn.PLT_FIGSIZE] = (20, 16)
mutation_plot.plotSiglvls(max_siglvl=0.05, parms=parms, is_time_lag=True)
```
